```python
from nilearn import plotting as niplt
from pathlib import Path
```

```python
result_imgs = {}
for folder in Path('.').iterdir():
    if folder.is_dir() and not folder.name.endswith('.ipynb'):
        result_imgs[folder.name] = folder.absolute() / 'z_corr-FDR_method-indep.nii.gz'
```

```python
?niplt.plot_stat_map
```

```python
# Plot all with niplt.plot_stat_map
order = ['all_analyses', 'affiliation_attachment', 'social_communication', 'perception_self', 'perception_others']
for name in order:
    img = str(result_imgs[name])
    display = niplt.plot_stat_map(
        img,
        display_mode='z',
        threshold=1.65,
        vmin=1.65,
        cut_coords=[-42, -25, -8, 10, 28, 45, 62],
        title=name,
        colorbar=True,
        cmap='YlOrRd'
    )
```
